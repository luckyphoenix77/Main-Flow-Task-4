```python
dt["trending_date"] = dt["trending_date"].apply(lambda x : datetime.datetime.strptime(x, "%y.%d.%m"))
```

```python
dt.head()
```

```python
dt["publish_time"] = pd.to_datetime(dt["publish_time"])
dt.head()
```

```python
dt["publish_month"] = dt["publish_time"].dt.month
dt["publish_day"] = dt["publish_time"].dt.day
dt["publish_hour"] = dt["publish_time"].dt.hour
```

```python
print(sorted(dt["category_id"].unique()))
```

```python
dt["category_name"] = np.nan
dt.loc[(dt["category_id"] == 1), "category_name"] = "Films and Animation"
dt.loc[(dt["category_id"] == 2), "category_name"] = "Autos and Vehicles"
dt.loc[(dt["category_id"] == 10), "category_name"] = "Music"
dt.loc[(dt["category_id"] == 15), "category_name"] = "Pets and animals"
dt.loc[(dt["category_id"] == 17), "category_name"] = "Sports"
dt.loc[(dt["category_id"] == 19), "category_name"] = "Travel and Events"
dt.loc[(dt["category_id"] == 20), "category_name"] = "Gaming"
dt.loc[(dt["category_id"] == 22), "category_name"] = "People and Blogs"
dt.loc[(dt["category_id"] == 23), "category_name"] = "Comedy"
dt.loc[(dt["category_id"] == 24), "category_name"] = "Entertainment"
dt.loc[(dt["category_id"] == 25), "category_name"] = "News and politics"
dt.loc[(dt["category_id"] == 26), "category_name"] = "How to Style"
dt.loc[(dt["category_id"] == 27), "category_name"] = "Education"
dt.loc[(dt["category_id"] == 28), "category_name"] = "Science and Technology"
dt.loc[(dt["category_id"] == 29), "category_name"] = "Non Profit Organisation"
dt.loc[(dt["category_id"] == 43), "category_name"] = "Shows"
```

```python
dt.head()
```

```python
dt["year"] = dt["publish_time"].dt.year
yearly_count = dt.groupby("year")["video_id"].count()

#bar chart
yearly_count.plot(kind='bar', xlabel = 'year', ylabel = 'Total publish count', title = 'Total publish video per year')
plt.show()
```

```python
yearly_views = dt.groupby('year')['views'].sum()
yearly_views.plot(kind='bar', xlabel = 'year', ylabel = 'total views', title = 'Total views per year')
plt.xticks(rotation = 0)
plt.tight_layout()

plt.show()
```

```python
category_views = dt.groupby('category_name')['views'].sum().reset_index()
top_categories = category_views.sort_values(by = 'views', ascending = False).head(5)
plt.bar(top_categories['category_name'],top_categories['views'])
plt.xlabel('category_name', fontsize=12)
plt.ylabel('total_views', fontsize=12)
plt.title('Top 5 categories',fontsize=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = "category_name", data = dt, order=dt['category_name'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Video Count by Category')
plt.show()
```

```python
videos_per_hr = dt['publish_hour'].value_counts().sort_index()

plt.figure(figsize=(12,6))
sns.barplot(x=videos_per_hr.index, y=videos_per_hr.values, palette='rocket')
plt.title('Number of Videos Published per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Views')
plt.xticks(rotation=45)
plt.show()
```

```python
dt['publish_time'] = pd.to_datetime(dt['publish_time'])
dt['publish_date'] = dt['publish_time'].dt.date
video_count_by_date = dt.groupby('publish_date').size()
plt.figure(figsize=(12,6))
sns.lineplot(data=video_count_by_date)
plt.title("Video Published Over Time")
plt.xlabel('Publish Date')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=dt, x='views', y = 'likes')
plt.title("Views vs Likes")
plt.xlabel=( 'views')
plt.ylabel= ('likes')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplots_adjust(wspace=0.2, hspace=0.4, top = 0.9)
plt.subplot(2,2,1)
g = sns.countplot(x= 'comments_disabled', data=dt)
g.set_title("Comments Disabled", fontsize=16)
plt.subplot(2,2,2)
g1 = sns.countplot(x='ratings_disabled',data=dt)
g1.set_title("Ratings Disabled", fontsize=16)
plt.subplot(2,2,3)
g2 = sns.countplot(x='video_error_or_removed', data = dt)
g2.set_title("Video Error or Removed", fontsize=16)
plt.show()
```
